`raster-vision`把遥感深度学习分为四个步骤：
1. Read geospatial data， 准备数据（矢量、影像）
2. Train a model，训练模型
3. Make predictions，模型推理预测
4. Write predictions (as geospatial data)，将识别结果转为矢量数据

# Read geospatial data
数据这块就需要栅格和矢量数据，完整流程`raster-vision`画了一个图，我放到这里了

这里面有几个概念解释下：
1. `LabelSource`，就是基于矢量数据形成的样本数据，然后转为栅格数据，作为训练数据
2. `Scene`， 就是`LabelSource`和`RasterSource`的组合
3. `GeoDataset`，类似于`torch.DataLoader`

# Training a model
`raster-vision`提供了一个`pytorch_learner`插件，封装了整个训练过程的参数，只需要传入`LearnerConfig`，设置数据集、模型、损失、优化器等就可以进行自动化训练，也可以传入自己的模型、数据集等。

# Making predictions and saving them
在模型训练完成后，就可以用它来推理。流程是：
1. 从模型实例化一个`Learner`
2. 将需要预测的影像构建为`GeoDataset`
3. 将预测转换为`Labels`(栅格转矢量，并且带坐标系信息)
4. 将标签保存到文件
